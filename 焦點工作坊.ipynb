##### **range()**  

**產生一個有序項目**（袋子）

````python
range(1, 11)
````

````python
list(range(1, 11))
````

````python
# 第一個數字包含，第二個數字"不包含"
# 也就是從 1 ~ 10，一共執行 10 次
# 之所以不包含第二個數字，是為了方便計算: 11-1 剛好等於 10 (執行次數) (就是國小的種樹問題)
for i in range(1, 11):
	print(i)
````

可以理解成：

````python
i = 1
print(i)

i = 2
print(i)

# 持續到
i = 10
print(i)
````

### while

- **<font color=red>記得縮排，按 tab 鍵</font>**

```
while 某個條件:
	# 做某件事
	# 執行完一次後，回到最上方，做條件判斷
```

##### 無限迴圈

**<font color=red>不要使用，務必要寫條件</font>**

````python
# 記得要手動中止
while True:
    print("Yes!")
````

##### 印出 1 ~ 10

````python
x = 0
while x < 10:
	x = x + 1
	print(x)
````

##### 差異

- for 迴圈，有固定次數
- while 迴圈，沒有固定次數

````python
# 固定執行 10 次迴圈
for i in range(1, 11):
	print(i)
````

````python
# 滿足 x < 10 ，則執行迴圈
x = 0
while x < 10:
	x = x + 1
	print(x)
````

##### 參考解答

還記得「當程式碼需要不停地複製貼上時，代表有優化的空間」這個觀念嗎？  
有兩段程式碼重複了，這又是一個完善程式碼的契機  
至於如何改善會需要使用 "function" 這個功能，這部分就留給各位回去好好研究了    

````python
# 程式隨機抽取數字
from random import randint
answer = randint(1, 10)  # 從 1~10 隨機取值
# print(answer)

# 使用者輸入一個值，並嘗試轉換成整數
try:
    your_answer = input("輸入一個整數: ")
    your_answer = int(your_answer)  # 這行可能出錯
except ValueError:
    print("請輸入整數")  # 如果出錯，代表輸入的值不是整數

# 當沒猜到正確答案時，執行迴圈
while your_answer != answer:
    # 判斷是否為整數
    if type(your_answer) == int:
	    # 比正確答案大
        if your_answer > answer:
            print("小一點")  # 印出訊息
	    # 比正確答案小
        elif your_answer < answer:
            print("大一點")  # 印出訊息

    # 使用者再次輸入一個值，並嘗試轉換成整數
    try:
        your_answer = input("輸入一個整數: ")
        your_answer = int(your_answer)
    except ValueError:
        print("請輸入整數")

print("猜對了")
````

或者，也可以用其他方式來優化，避免程式碼重複  
寫程式本身沒有絕對的對錯，能解決問題的都是好方法  

````python
# 程式隨機抽取數字
from random import randint
answer = randint(1, 10)  # 從 1~10 隨機取值
# print(answer)

# your_answer 的默認值，必然會進入迴圈
your_answer = ""

# 當沒猜到正確答案時，執行迴圈
while your_answer != answer:
    # 使用者輸入一個值
    your_answer = input("輸入一個整數: ")

    # 檢查輸入是否為 十進位 數字
    if your_answer.isdecimal():
        # 轉成整數
        your_answer = int(your_answer)

        # 比正確答案大
        if your_answer > answer:
            print("小一點")  # 印出訊息
        # 比正確答案小
        elif your_answer < answer:
            print("大一點")  # 印出訊息
    else:
        print("請輸入整數")

print("猜對了")
````
